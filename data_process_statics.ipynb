# Data processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data_path = 'laptop.csv'
laptop_data = pd.read_csv(data_path)

# Calculate average prices
average_prices = laptop_data.groupby('Brand')['Price'].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x=average_prices.values, y=average_prices.index, palette='viridis')
plt.title('Average Laptop Prices by Brand')
plt.xlabel('Average Price')
plt.ylabel('Brand')

# Display the plot
plt.show()
```

```python
# Create a histogram 
plt.figure(figsize=(10, 6))
sns.histplot(laptop_data['Price'], bins=30, color='skyblue')
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np

# Convert RAM column 
laptop_data['RAM_GB'] = laptop_data['RAM'].str.extract('(\d+)').astype(float)

# Recreating
colors = plt.cm.viridis(np.linspace(0, 1, len(laptop_data)))

# Create a scatter plot 
plt.figure(figsize=(10, 6))
plt.scatter(laptop_data['RAM_GB'], laptop_data['Price'], color=colors, edgecolor='black')
plt.title('RAM vs. Price of Laptops (Colorful)')
plt.xlabel('RAM (GB)')
plt.ylabel('Price (in USD)')
plt.grid(True)
plt.show()
```
