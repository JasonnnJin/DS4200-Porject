# Data processing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data_path = 'laptop.csv'
laptop_data = pd.read_csv(data_path)

# Calculate average prices
average_prices = laptop_data.groupby('Brand')['Price'].mean().sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(14, 8))
bar_plot = plt.barh(average_prices.index, average_prices.values, color='steelblue')
plt.title('Average Laptop Prices by Brand')
plt.xlabel('Average Price')
plt.ylabel('Brand')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Code for histogram
plt.figure(figsize=(10, 6))
sns.histplot(laptop_data['Price'], bins=30, color='steelblue')
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# RAM vs Price
import numpy as np

# Convert RAM column and create scatter with regression line
laptop_data['RAM_GB'] = laptop_data['RAM'].str.extract('(\d+)').astype(float)

# Create scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='RAM_GB', y='Price', data=laptop_data, 
           scatter_kws={'alpha':0.5}, 
           line_kws={'color': 'red'},
           color='steelblue')
plt.title('RAM vs Price with Trend Line')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.grid(True)
plt.tight_layout()
plt.show()
```
